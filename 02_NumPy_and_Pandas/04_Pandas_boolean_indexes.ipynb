## 1. Векторизированные операции
Помните ли вы векторные операции, с которыми мы сталкивались в библиотеке NumPy? Векторные операции позволяют применять операции сразу к нескольким точкам данных, что не только улучшает производительность нашего кода, но и позволяет нам писать код быстрее.

Поскольку pandas является расширением NumPy, он также поддерживает векторные операции. Как и в NumPy, мы можем использовать с **сериями** любые стандартные числовые операторы Python, включая:
- ``Сложение``: вектор_a + вектор_b
- ``Вычитание``: вектор_a - вектор_b
- ``Умножение``: вектор_a * вектор_b (не связано с умножением векторов в линейной алгебре)
- ``Деление``: вектор_a / вектор_b

## 3. Метод Describe в сериях
В этой главе мы изучим еще один метод под названием **Series.describe()**. Этот [метод](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.describe.html) показывает нам различную информацию, например, сколько ненулевых значений
содержится в серии, среднее значение, минимум, максимум и другие статистические данные. Давайте посмотрим, как мы можем использовать этот метод:

`````python
assets = f500["assets"]
print(assets.describe())
`````

Вы можете видеть, что значения в приведенном выше сегменте кода отображаются в **E- нотации**, разновидности [научной нотации](https://en.wikipedia.org/wiki/Scientific_notation).

Когда мы используем **describe()** для столбца, который содержит нечисловые значения, мы получим несколько другую информацию по статистикам нежели как показано в предыдущем примере:

`````python
country = f500["country"]
print(country.describe())
`````

Первая строка содержит информацию о **count**, такая же, как и для числовых столбцов, показывает нам количество ненулевых значений. Остальные три статистики описаны ниже:
- ``unique``: Количество уникальных значений в серии
- ``top``: Наиболее часто встречающееся значение в серии
- ``freq``: Частота наиболее часто встречающегося значения

## 4. Цепочка вызова функций
Цепочка вызова функций - это обычный синтаксис для вызова нескольких функций, идущих друг за другом, в объектно-ориентированных языках программирования. Каждая функция возвращает объект, что позволяет объединять вызовы в единый оператор, не требуя переменных для хранения промежуточных результатов ([Википедия](https://en.wikipedia.org/wiki/Method_chaining)).

Давайте рассмотрим на примере как это выглядит:

Без цепочки вызова функций

````python
countries = f500["country"]
countries_counts = countries.value_counts()
````

С помощью цепочки вызова функций

````python
countries_counts = f500["country"].value_counts()
````

С этого момента мы будем стараться использовать в коде все больше и больше цепочек вызова функций. При написании кода всегда оценивайте, не затруднит ли такой вызов чтение вашего кода. Если да, то всегда предпочтительнее разбить код на несколько строк.

## 5. Методы исследования датафреймов
Поскольку серии и датафреймы - это два разных объекта, они имеют свои собственные уникальные функции. Однако у них также есть функции с одинаковыми именами, которые ведут себя схожим образом. Ниже приведены некоторые примеры:
- `Series.max()` и `DataFrame.max()`
- `Series.min()` и `DataFrame.min()`
- `Series.mean()` и `DataFrame.mean()`
- `Series.median()` и `DataFrame.median()`
- `Series.mode()` и `DataFrame.mode()`
- `Series.sum()` и `DataFrame.sum()`

> В отличие от серий, методы dataframe требуют параметра axis(оси x или y), чтобы знать, по какой оси производить вычисления. Вы можете использовать целые числа для обозначения первой и второй оси. Методы dataframe Pandas также принимают строки ``index`` и ``columns`` в качестве параметра axis:


````python
# Когда мы пытаемся вычислить по оси x - строк
DataFrame.method(axis = 0)
# или
Dataframe.method(axis = "index")
````

````python
# Когда мы пытаемся вычислить по оси y - столбцов
DataFrame.method(axis = 1)
# или
Dataframe.method(axis = "column")
````

Для более конкретного примера, если мы хотим найти медиану для **revenue** и **profit** столбцов в нашем наборе данных, мы можем сделать следующее:

`````python
medians = f500[["revenues", "profits"]].median(axis=0)
# можно также использовать .median(axis="index")
print(medians)
`````

## 8. Использование булевой индексации с объектами pandas
Чтобы изменить много значений одновременно, мы рекомендуем использовать **булеву индексацию**.

Давайте рассмотрим пример датафрейма ниже. Это датафрейм о людях и их любимых номерах:

`````python
import pandas as pd

d = {'name': ["Kyllie", "Rahul", "Michael", "Sarah"], 'num': [12, 8, 5, 8]}

df = pd.DataFrame(data=d)
df
`````

Если мы хотим проверить, у кого из людей любимое число 8, мы можем сначала выполнить векторную булеву операцию, которая выдаст булеву серию:

`````python
num_bool = df["num"] == 8
num_bool
`````

Мы воспользуемся получившейся серии для индексации всего набора данных, в результате чего у нас останутся строки, соответствующие только тем людям, чье любимое число - 8:

`````python
result = df[num_bool]
result
`````

`````python
# или так
result = df.loc[num_bool,"name"]
result
`````

## 10. Использование булевых серий для присвоения значений
В этой главе мы рассмотрим, как можно объединить присвоение и булеву индексацию в pandas.

В следующем примере мы изменим значения 'Motor Vehicles & Parts' в столбце **sector** на 'Motor Vehicles and Parts' - т.е. мы изменим амперсанд **(&)** на **and**.

`````python
# Сначала мы создаем булеву серию, сравнивая значения в столбце sector
ampersand_bool = f500["sector"] == "Motor Vehicles & Parts"

# Далее мы используем эту булеву серию и столбец "sector" для выполнения задания
f500.loc[ampersand_bool,"sector"] = "Motor Vehicles and Parts"
`````

`````python
# Проверяем результат
f500.loc[ampersand_bool,"sector"]
`````

Мы можем сделать то же самое, убрать промежуточный шаг создания булевой серии и объединить все в одну строку, вот так:

`````python
f500.loc[f500["sector"] == "Motor Vehicles & Parts","sector"] = "Motor Vehicles and Parts"
`````

После того как мы научились комбнировать функции в одну строку, мы можем использовать эту технику и заменить значения в столбце **previous_rank**.
