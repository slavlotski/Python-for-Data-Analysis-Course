## Механики рисования в Matplotlib
Создание рисунка в matplotlib схоже с рисованием в реальной жизни. Так художнику нужно взять основу (холст или бумагу), инструменты (кисти или карандаши), иметь представление о будущем рисунке (что именно он будет рисовать) и, наконец, выполнить всё это и нарисовать рисунок деталь за деталью. В matplotlib все эти этапы также существуют, и в качестве художника-исполнителя здесь выступает сама библиотека. От пользователя требуется управлять действиями художника-matplotlib, определяя что именно он должен нарисовать и какими инструментами

## 3.1 Линейный график
Давайте создадим объект Figure и 3 рисунка линейного графика, используя общую функцию .plot() в стиле MATLAB.

Используя **plt.subplot()**, мы создаем 3 рисунка, где задано следующее:
- количество строк - первый параметр
- количество столбцов - второй параметр
- индекс - третий параметр (это положение рисунка в сетке,
индекс всегда начинается с 1 в левом верхнем углу, индекс увеличивается на 1 вправо)

```python
# Пример данных для рисования
data = np.arange(0.0, 50, 0.01)
# Синус
s = np.sin(data)
# Косинус
c = np.cos(data)
# Пример данных для рисования
tg = np.arange(0.0, 1.5, 0.01)
# Тангенс
t = np.tan(tg)

# Создаем объект Figure
plt.figure(figsize = (15,9))

# Создаем первый мультиоконный рисунок
plt.subplot(2,2,1)               # 2 строки, 2 колонки, и 1-ый индекс
plt.plot(data,s,                 # Вызов функции plot на первом рисунке
         color = 'red')
plt.title('The first subplot')   # Название первого рисунка

# Создаем второй рисунок
plt.subplot(2,2,2)               # 2 строки, 2 колонки, и 2-ой индекс рисунка
plt.plot(data,c,                 # Вызов функции plot на втором рисунке
         color = 'yellow')
plt.title('The second subplot')

# Создаем третий рисунок
plt.subplot(2,2,3)               # 2 строки, 2 колонки, и 3-ий индекс рисунка
plt.plot(tg,t,                   # Вызов функции plot на третьем рисунке
        'green',
         alpha = 0.5)
plt.title('The third subplot')
plt.suptitle('Subplots');        # Заголовок всего объекта Figure
```

Теперь давайте воссоздадим рисунок, используя объектно-ориентированный интерфейс Matplotlib. При таком подходе вам не нужно беспокоиться о том, какая ось (Axis) активна в данный момент. Функции построения применяются явно к объекту Axes, который вы выбираете.

Если у вас есть сетка у мультиоконного рисунка, состоящая из двух рисунков, вы можете получить доступ к конкретной оси, индексируя ее в соответствии с ее положением. В этом случае индексация начинается с нуля.

```python
# Создаем объект Figure и группу рисунков
fig1, ax = plt.subplots(1,2)

# Первый рисунок доступен по ах[0]
ax[0].plot(data, s, color = 'red')

# Второй рисунок доступен по ах[1]
ax[1].plot(data, c, color = 'yellow')

# Удаляем разметку для оси x (иногда это уместно)
ax[1].set_xticks([])

# Регулируем отступы между рисунками
plt.tight_layout()
```

```python
# Создаем объект Figure и 3 рисунка
fig, ax = plt.subplots(2,2, figsize = (15,9))

# Первый рисунок с красной линией будет расположен на 1 строке и 1 столбце
ax[0,0].plot(data, s, color = 'red')

# Далее мы идем направо к рисунку с желтой линией в 1 строку и 2 столбец
ax[0,1].plot(data, c, color = 'yellow')

# Далье идет рисунок с зеленой линией вниз в нижний угол, 2 строка и 1 стобец
ax[1,0].plot(tg, t, color = 'green');

# Для 2 строки 2 столбца мы ничего не указали, поэтому последний рисунок пустой
```

На выходе **plt.subplots(2,2)** - это двумерный массив, поэтому обьекты Axes задаются их расположением в соответствующей строке и столбце.

Вы сами решаете, какой подход к построению графиков выбрать. Но общая рекомендация такова: Используйте MATLAB-подобный стиль рисования, когда вы создаете только простые графики. Для сложных графиков сболее чем одним рисунком или добавлением других компонентов используйте объектно-ориентированный подход к построению графиков, который дает вам полный контроль над объектом Axes.

## 3.2 Гистограмма
Для создания гистограммы можно использовать функцию **hist()**,в которой можно указать набор параметров. См. [документацию](https://matplotlib.org/3.3.3/api/_as_gen/matplotlib.pyplot.hist.html)

Мы используем **plt.subplots()** для создания объекта Figure и сетки рисунков (объекты Axes), где мы задаем 1 строку и 2 столбца, так что наши рисунки будут созданы рядом друг с другом. После этого мы получим доступ к каждому из объектов Axes по их индексам и сможем создавать графики.

```python
# Для демонстрации мы создадим случайную выборку из нормального распределения
# с помощью numpy
np.random.seed(42)
x_var2 = np.random.normal(size = 150)
```

```python
# Создаем объект Figure и Axes с указанием расположения рисунков с нужным нам размером
fig, ax = plt.subplots(1,2, figsize = (13,9))
# Создаем гистограмму на первом рисунке
# Обратите внимание, что индекс начинается с 0
ax[0].hist(x_var2, bins = 15, color = 'red',  histtype = 'step')
ax[0].set_xlabel('Values')
ax[0].set_ylabel('Count')
ax[0].set_title('The distribution of data points')

# Создаем гистограмму плотности во втором рисунке
ax[1].hist(x_var2, bins = 10, density = True)
ax[1].set_xlabel('Values')
ax[1].set_ylabel('Density')
ax[1].set_title('The density of data points')

# Настраиваем отступ между краями рисунков
plt.tight_layout(pad = 6)
# Добавляем общий заголовок
# Настраиваем положение заголовка, используя значения x и y
# Настраиваем размер шрифта
fig.suptitle('Histograms of x_var', x = 0.5, y = 0.97, fontsize = 20);
```

## 3.3 Ящик с усами
Вы можете создать ящик с усами с помощью функции **plt.boxplot()**. см.
документацию [здесь](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html). Существует несколько вариантов настройки вывода, например, настройка элементов графика ящика с усами.

```python
# Создаем объект Figure
fig, ax = plt.subplots(figsize = (13,9))
# Создаем график ящик с усами с помощью переменной x_var
box_properties = dict(color = 'orange', linestyle = '--')           # Свойства для ящика
wh_properties = dict(linestyle = '-.',linewidth = 1.5)              # Свойства для усов
median_line = dict(color = 'red', linewidth = 2)                    # Свойства для линии медианы
mean_line = dict(color = 'green', linewidth = 3, linestyle = '-.')  # Свойства для среднего значения
ax.boxplot(x_var2,
           showmeans = True,                 # Показываем среднее значение на графике
           meanline = True,                  # Отображаем среднее значение в виде линии
           meanprops = mean_line,            # Свойства для среднего значения
           boxprops = box_properties,        # Свойства для ящика
           whiskerprops = wh_properties,     # Свойства для усов
           medianprops =  median_line);      # Свойства для медианы
```

##  3.4 Столбчатая гистограмма
Для визуализации и сравнения категорий переменных можно построить столбчатую
гистограмму. Количество точек данных, относящихся к определенной категории,
будет изображено в виде столбика.

Функция **bar** в Matplotlib принимает несколько параметров, таких как:
- x: координаты столбцов
- y: высота столбиков, другими словами, количество точек данных

Другие параметры включают установку ширины полос или их выравнивание. Можете обратиться к документации [здесь](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.bar.html) для большей информации.

Теперь мы построим столбчатую гистограмму, на которой изобразим максимальную скорость животных и сравним ее с максимальной скоростью Усэйна Болта на чемпионате мира 2009 года, которая составляет 44,72 км/ч.

```python
animals = ['Horse','Pig','Cheetah','Coyote','Ostrich','Black Mamba']
top_speed = [88,17.7,120,69,70,32.2]
```

```python
# Создаем объект Figure
fig,ax = plt.subplots(figsize = (9,7))
# Поэксперементируйте, как ширина полос может меняться в зависимости от параметра width
ax.bar(animals, top_speed, width = 0.5)
# В конце концов, добавьте параметр 'align = 'edge'' к приведенной выше строке кода, чтобы увидеть,
# как изменится выравнивание столбцов

# Устанавливаем название оси x
plt.xticks(animals)
# Устанавливаем угол наклона оси x для каждой категории
plt.xticks(rotation = 45)
# Устанавливаем заголовок
plt.title('Comparing top speed', fontsize = 16)
plt.xlabel('Animals')
plt.ylabel('Top speed (km/h)')

# Добавляем горизонтальную линию для изображения максимальной скорости Усэйна Болта
# вместе со стрелкой, указывающей на эту линию
ax.axhline(y = 44.72, linewidth = 2, alpha = 0.7, color = 'red')
ax.annotate('Usain Bolt \n 44.72 km/h', xy = (4.9,45), xytext=(4.9,57), ha = 'center', fontsize = 14,
           arrowprops = dict(arrowstyle = '->', color = 'red'))
;
# Убираем рамку графика сверху и справа
ax.spines['top'].set(visible = False)
ax.spines['right'].set(visible = False);
```
