## 3. Синтаксис в Python
Чуть выше мы получили **синтаксическую ошибку**, это потому что языки программирования такие как Python, Java, C++ имеют свои синтаксические правила. Каждая строчка кода должна соответствовать этим правилам. 

Вы можете сравнить эти синтаксические правила с грамматикой в наших естественных языках. Если мы хотим передать сообщение, нам стоит уважать и следовать этим синтактическим правилам, чтобы донести наше сообщение в понятной форме. Например, люди поймут следующее сообщение "Data Science - это супер круто", но не поймут такое "Science круто это супер Data". Аналогично и для Python, он не поймет следующую строчку кода `print(55+5)` `print(300-8)` `print(21+67)`, потому что это синтаксически неправильно. 

**С ошибками в языках программирования вы будете встречаться довольно часто**.

## 5. Арифметические операции

В предыдущих секциях мы выполняли операции сложения и вычетания. Для умножения нам необходимо использовать оператор `*`. Чтобы перемножить 5 на 2 мы должны напечатать следующее: 

```python
5 * 2
```

Для деления необходимо использовать оператор `/`. Чтобы разделить 10 на 5:

```python
10 / 5 
```

Для возведения числа в степень необходимо использовать `**`. Например, возведем 2 в степень 3:

```python
2 ** 3
```

Арифметические операции в Python следуют той же очередности выполнения, что и в математике, сначала выполняется выражение в скобках, следом идет возведение в степень, деление, умножение и наконец сложение с вычитанием.

## Арифметические операторы в Python

| Оператор |  Название |  Пример |
|------------|:------:|----------:|
| + | Сложение | x + y| 
| - | Вычитание| x - y| 
| * | Умножение | x * y| 
| / | Деление | x / y| 
| ** | Возведение в степень | x ** y| 
| % | Остаток от деления | x % y|
